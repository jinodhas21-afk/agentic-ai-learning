```python
from openai import OpenAI
from IPython import display
client = OpenAI()

response = client.images.generate(
    model="dall-e-2",
    prompt="a white siamese cat",
    n=1,
    size="1024x1024"
)
url = response.data[0].url
display.Image(url=url)
```
